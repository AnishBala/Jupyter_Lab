Read excel data and skip first 6 rows to get the correct line for headers

```python
df = pd.read_excel(
    "Signal & Light Inventory - Signal Inventory D3 2024-08-13.xlsx", skiprows=6
)
```

select as built options "3H3304" and "3H330". Data under "district" and "county" headers do not merge properly so the ffill function has to be used on these headers. Print a data file with only the specified as built options and the corrected merged data.

```python
options = ["3H3304", "3H330"]
df[["District", "County"]] = df[["District", "County"]].ffill()
rslt_df = df[df["As Built"].isin(options)]
rslt_df
```

one way to count the number of as builts per signal type using a for loop and a dictionary

```python
sigtype_df = rslt_df.groupby(["Signal Type"])
signal_type_dict= sigtype_df.groups
for key in signal_type_dict:
    x = signal_type_dict.get(key)
    count = len(x)
    print(key + " count is " + str(count))
```

another way to count using the signal type grouping and using the size funciton

```python
type_counts = rslt_df.groupby(["Signal Type"]).size()
type_counts 
```
